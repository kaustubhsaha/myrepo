```python
players = {}
for (i, player) in enumerate(Match(match_id).team_1_players):
    short_name = player['mobile_name']
    full_name = player['card_long']
    players[short_name] = full_name
for (i, player) in enumerate(Match(match_id).team_2_players):
    short_name = player['mobile_name']
    full_name = player['card_long']
    players[short_name] = full_name
```

```python
dismissals = batting['out'][batting['out'] != 'not out']
```

```python
run_out_fielders = []
catch_fielders = []
stumping_wk = []

for d in dismissals:
    if ('run out' in d):
        player_short_name = d[d.find("(")+1:d.find(")")]
        if ("/" in player_short_name):
            player_short_name = player_short_name[0:player_short_name.find("/")]
            if (player_short_name.startswith("wk")):
                player_short_name = player_short_name[2:]
        run_out_fielders.append(players[player_short_name])
    elif (d.startswith('c  b')):
        player_short_name = d.split(" ")[3]
        if ("/" in player_short_name):
            player_short_name = player_short_name[0:player_short_name.find("/")]
        catch_fielders.append(players[player_short_name])
    elif (d.startswith('c ')):
        player_short_name = d.split(" ")[1]
        if ("/" in player_short_name):
            player_short_name = player_short_name[0:player_short_name.find("/")]
        catch_fielders.append(players[player_short_name])
    elif (d.startswith('st ')):
        player_short_name = d.split(" ")[1]
        if (player_short_name.startswith("wk")):
            player_short_name = player_short_name[2:]
        stumping_wk.append(players[player_short_name])
```

```python
fielding = pd.DataFrame()

for p in run_out_fielders:
    fielding = fielding.append(pd.Series([p, 'RunOut'], index=["Name","Type"]), ignore_index=True)
for p in catch_fielders:
    fielding = fielding.append(pd.Series([p, 'Catch'], index=["Name","Type"]), ignore_index=True)
for p in stumping_wk:
    fielding = fielding.append(pd.Series([p, 'Stumping'], index=["Name","Type"]), ignore_index=True)

fielding = fielding.set_index('Name')
```

Let's load the dataframe again (and not treat the first row as header)

Lets put meaningful labels to the data columns

```python
data = pd.DataFrame(raw_data.values, columns=['O-Rings at risk','O Rings at Thermal distress','Launch Temperature','Leak-check pressure','Temporal order'])
```

```python
display(HTML(data.to_html()))
```

Clearly we can use Temporal order as the index and drop the default montonic index

```python
data.set_index('Temporal order', inplace=True)
```

```python
display(HTML(data.to_html()))
```

Let's try building a predictive model to observe the O-ring degradation at different temperatures. <br>
This model could come in handy in forecasting and in order to predict what an optimal temperature would be to support these O-rings.

The inputs to this model will be :
1. Launch Temperature
2. Leak-check Pressure

The output of the model will be
1. O Rings at Thermal distress

Now. for multiple regression, its important that there is no multicollinearity ie the input variables should be independent of each other and must not have a high correlation among themselves

Let's check if the data satisfies that criteria by generating a correlation matrix

```python
plt.cla()
plt.clf()
corr_matrix = data[['O Rings at Thermal distress','Launch Temperature','Leak-check pressure']].corr().round(2)

fig = plt.figure(figsize = (10,10))
sns.heatmap(corr_matrix, annot=True, center=0, square=True, cmap="summer")
plt.title("Correlation Matrix")
plt.show()
```

Clearly the independent variables in our regression model - launch Temperature and Leak Check pressure are nearly independent (correlation of 0.04)

There is also some level of correlation between out output variable (O Rings at Thermal distress) and the two independent variables (0.51 and 0.28) indicating a potential cause-effect relationship

So at a high level, our model looks good

 Linear regression models are highly sensitive to outliers. So, lets inspect the data and check if we have any outliers

```python
plt.cla()
plt.clf()
plt.scatter(data['Launch Temperature'],data['O Rings at Thermal distress'],
         marker='o', 
         edgecolor='b',
         facecolor='none',
         alpha=0.85 )
plt.xlabel('Launch Temperature')
plt.ylabel('O Rings at Thermal distres')
plt.title("O Rings at Thermal distres vs Launch Temperature")
plt.show()
```

```python
plt.cla()
plt.clf()
plt.scatter(data['Leak-check pressure'],data['O Rings at Thermal distress'],
         marker='o', 
         edgecolor='b',
         facecolor='none',
         alpha=0.85 )
plt.xlabel('Leak-check pressure')
plt.ylabel('O Rings at Thermal distres')
plt.title("O Rings at Thermal distres vs Leak-check pressure")
plt.show()
```

Clearly there are no outliers and the number of O rings at Thermal distress always varies in 0-2 range

We'll utilize the ols tool from statsmodels package to generate the regression model. 
The ols tool gives us the intercept, slope, and as well as the R-squared value — a handy tool to help us see whether our results are significant.

```python
model_data = data.rename(columns= {'O Rings at Thermal distress' : 'Y', 'Launch Temperature' : 'X0', 'Leak-check pressure' : 'X1'})
display(HTML(model_data.head().to_html()))
```

```python
model = sm.ols('Y ~ X0 + X1',data = model_data)
```

```python
results = model.fit()
results.summary()
```

So, we can represent the data using the regression equation :<br>
   <b> Y = 3.3298 - 0.0487*X0 + 0.0029*X1 </b><br>
   <i>where,</i><br>
    Y -> O Rings at Thermal distress<br>
    X0 -> Launch Temperature<br>
    X1 -> Leak-check pressure

Now we need to predict how many O-rings does the model predict will show erosion or blowby when the temperature is 31 degrees F. We don’t know the pressure (PSI) we should be making the prediction for. Therefore, it is beneficial to observe for all the pressures we have been given in our data (0, 50, 100, 200).

Lets try to get the get the values using the derived equation

```python
def get_o_ring_distress(temp, pressure):
    return 3.3298 - 0.0487*temp + 0.0029*pressure

print("Predicted number of O rings under distress at 0 PSI and 31 degrees F :", get_o_ring_distress(31,0))
print("Predicted number of O rings under distress at 50 PSI and 31 degrees F :", get_o_ring_distress(31,50))
print("Predicted number of O rings under distress at 100 PSI and 31 degrees F :", get_o_ring_distress(31,100))
print("Predicted number of O rings under distress at 200 PSI and 31 degrees F :", get_o_ring_distress(31,200))
```

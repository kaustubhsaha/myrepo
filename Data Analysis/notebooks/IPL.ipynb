```python
IPL_DATASET_MATCH = r"../data/IPL/Match.csv"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(IPL_DATASET_MATCH)
df.head()
```

```python
def toss_winner_won (row):
    if (row['Toss_Winner'] == row['match_winner']) : return 1
    else : return 0
df['Toss Decision']=df['Toss_Name'].str.upper()
df['Toss_Winner_Won']=df.apply(lambda p : toss_winner_won(p), axis = 1)
grouped_by_decision = df.groupby('Toss Decision').agg(np.mean)
grouped_by_decision['Toss_Winner_Won'].plot(kind = 'bar', x='Toss Decision', y='Win probability', title='Win probability vs Toss decision')
```

```python
grouped_by_decision_team = df.groupby(['Toss_Winner','Toss Decision']).agg(np.mean)
grouped_by_decision_team['Toss_Winner_Won'].plot(kind = 'bar', x='Toss Decision', y='Win probability', title='Win probability vs Toss decision', figsize=(12,2))
```

```python
grouped_by_decision_team = df.groupby(['Venue_Name','Toss Decision']).agg(np.mean)
grouped_by_decision_team['Toss_Winner_Won'].plot(kind = 'bar', x='Toss Decision', y='Win probability', title='Win probability vs Toss decision', figsize=(20,2))
```
